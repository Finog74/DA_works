####  Определение товарных категорий

## Сегментация покупателей на основе покупок. 

Сегментацию покупателей проведем с использованием k-means кластеризации.    
Разработаем и применим модель сегментации покупателей, которая учитывает широкий спектр параметров и автоматически определяет сегменты без жёстких ограничений. Это позволит более точно идентифицировать различные типы клиентов и разработать персонализированные маркетинговые стратегии для каждого сегмента.  
Основные сегменты клиентов которые можно выделить с помощью кластеризации k-means:  
- клиенты с высоким доходом
- потенциальные оптовики
- случайные покупатели
- лояльные клиенты        

Для проведения кластеризации покупателей методом k-means будем использовать следующие показатели:

**Средняя цена товара** Мжет быть основным параметром для группировки покупок по ценовым категориям.      
**Среднее количество товара** Количество купленного товара может помочь выявить закономерности в покупательском поведении.     
**Количество заказов** Для выявления лояльности покупателей.     
**Средняя стоимость заказа**

```python
# определим количество сегментов
features = ['quantity', 'price', 'revenue']
X = df[features]

X = X.fillna(X.mean()) 

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Методом локтя:
inertia = []
range_n_clusters = range(2, 11)  # Test a range of cluster numbers

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, inertia, marker='o')
plt.title('Метод Локтя для определения оптимального количества кластеров')
plt.xlabel('Количество кластеров')
plt.ylabel('Инерция')
plt.show()

# 2. Методом оценки силуэта:
silhouette_scores = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_scaled)
    silhouette_avg = silhouette_score(X_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot the Silhouette Score graph
plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.title('Метод силуэта для определения оптимального количества кластеров')
plt.xlabel('Количество кластеров')
plt.ylabel('Оценка силуэта')
plt.show()
```

 ## Формулировка и проверка статистических гипотез

Гипотеза 1:     
Нулевая гипотеза (H0): не существует статистически значимой разницы в среднем чеке между сегментами покупателей.          
Альтернативная гипотеза (H1): существует статистически значимая разница в среднем чеке между постоянными и эпизодическими покупателями.

```python
# 1. Проверка на нормальность распределения (Shapiro-Wilk)

cluster_revenues = [df['revenue'][df['cluster'] == i] for i in df['cluster'].unique()]

normality_results = {}
for i, revenue in enumerate(cluster_revenues):
    stat, p = stats.shapiro(revenue)
    normality_results[i] = {'statistic': stat, 'p-value': p}
    print(f"Кластер {i}: Shapiro-Wilk statistic={stat:.3f}, p-value={p:.3f}")

# Проверяем, что p-value больше alpha для каждого кластера.
alpha = 0.05
normal_clusters = all(normality_results[i]['p-value'] > alpha for i in normality_results)

if not normal_clusters:
    print("\nПредупреждение: Не все кластеры имеют нормальное распределение. Рассмотрите непараметрический тест (например, Kruskal-Wallis).")
else:
    print("\nВсе кластеры имеют нормальное распределение.")

# 2. Проведение теста ANOVA (если данные нормально распределены)
if normal_clusters:
    # 3. Проверка на достаточное количество кластеров
    if len(df['cluster'].unique()) < 2:
        print("Warning: Less than two clusters available. ANOVA cannot be performed.")
    else:
        # Выполняем тест ANOVA
        f_statistic, p_value = stats.f_oneway(*cluster_revenues)

        print("\nРезультаты ANOVA теста:")
        print(f"F-statistic: {f_statistic:.3f}")
        print(f"P-value: {p_value:.3f}")

        # 4. Учет множественных сравнений (Bonferroni correction)
        alpha_corrected = alpha / (len(df['cluster'].unique()) - 1)  # Bonferroni

        print(f"\nСкорректированный уровень значимости (Bonferroni): {alpha_corrected:.3f}")

        if p_value < alpha_corrected:
            print("\nОтвергаем нулевую гипотезу.")
            print("Существует статистически значимая разница в среднем чеке по крайней мере между двумя сегментами клиентов (с учетом поправки на множественные сравнения).")
        else:
            print("\nНе удается отвергнуть нулевую гипотезу.")
            print("Статистически значимой разницы в среднем чеке между клиентскими сегментами нет (с учетом поправки на множественные сравнения).")
else:
    print("\nТест ANOVA не может быть проведен из-за ненормального распределения. Рассмотрите непараметрический тест (например, Kruskal-Wallis).")


# 5. Альтернатива: Kruskal-Wallis test (если данные не нормально распределены)
if not normal_clusters:
    # Проверка на достаточное количество кластеров
    if len(df['cluster'].unique()) < 2:
        print("Warning: Less than two clusters available. Kruskal-Wallis cannot be performed.")
    else:
        # Выполняем тест Kruskal-Wallis
        h_statistic, p_value = stats.kruskal(*cluster_revenues)
        print("\nРезультаты теста Kruskal-Wallis:")
        print(f"Kruskal-Wallis statistic: {h_statistic:.3f}")
        print(f"P-value: {p_value:.3f}")

        # Поправка Бонферрони
        alpha_corrected = alpha / (len(df['cluster'].unique()) - 1)  # Bonferroni
        print(f"\nСкорректированный уровень значимости (Bonferroni): {alpha_corrected:.3f}")

        if p_value < alpha_corrected:
            print("\nОтвергаем нулевую гипотезу.")
            print("Существуют статистически значимые различия в среднем чеке между клиентскими сегментами (с учетом поправки на множественные сравнения).")
        else:
            print("\nНе удается отвергнуть нулевую гипотезу.")
            print("Статистически значимых различий в среднем чеке между клиентскими сегментами нет (с учетом поправки на множественные сравнения).")
```

Гипотеза 2    
Нулевая гипотеза (H0):
Нет статистически значимых различий в частоте покупок между постоянными и эпизодическими покупателями.
Альтернативная гипотеза (H1):     
Существуют статистически значимые различия в частоте покупок между постоянными и эпизодическими покупателями.

```python
# 1. Агрегация данных по клиентам
purchase_frequency = df.groupby('customer_id')['order_id'].count()
purchase_frequency = purchase_frequency.rename('purchase_frequency')

customer_data = df.groupby('customer_id').agg(
    avg_price=('price', 'mean'),
    avg_quantity=('quantity', 'mean'),
    num_orders=('order_id', 'nunique'),
    avg_order_value=('revenue', 'mean')
)
customer_data['purchase_frequency'] = purchase_frequency

cluster_map = df.groupby('customer_id')['cluster'].first().to_dict()
customer_data['cluster'] = customer_data.index.map(cluster_map)

# 2. Проверка на нормальность распределения (Shapiro-Wilk) для частоты покупок в каждом кластере
cluster_frequencies = [customer_data['purchase_frequency'][customer_data['cluster'] == i] for i in customer_data['cluster'].unique()]

normality_results = {}
for i, frequency in enumerate(cluster_frequencies):
    stat, p = stats.shapiro(frequency)
    normality_results[i] = {'statistic': stat, 'p-value': p}
    print(f"Кластер {i}: Shapiro-Wilk statistic={stat:.3f}, p-value={p:.3f}")

# Проверяем, что p-value больше alpha для каждого кластера.
alpha = 0.05
normal_clusters = all(normality_results[i]['p-value'] > alpha for i in normality_results)

if not normal_clusters:
    print("\nПредупреждение: Не все кластеры имеют нормальное распределение. Рассмотрите непараметрический тест (например, Kruskal-Wallis).")
else:
    print("\nВсе кластеры имеют нормальное распределение.")

# 3. Проведение теста ANOVA (если данные нормально распределены)
if normal_clusters:
    # Проверка на достаточное количество кластеров
    if len(customer_data['cluster'].unique()) < 2:
        print("Warning: Less than two clusters available. ANOVA cannot be performed.")
    else:
        # Выполняем тест ANOVA
        f_statistic, p_value = stats.f_oneway(*cluster_frequencies)

        print("\nРезультаты ANOVA теста:")
        print(f"F-statistic: {f_statistic:.3f}")
        print(f"P-value: {p_value:.3f}")

        # 4. Учет множественных сравнений (Bonferroni correction)
        alpha_corrected = alpha / (len(customer_data['cluster'].unique()) - 1)  # Bonferroni

        print(f"\nСкорректированный уровень значимости (Bonferroni): {alpha_corrected:.3f}")

        if p_value < alpha_corrected:
            print("\nОтвергаем нулевую гипотезу.")
            print("Существует статистически значимая разница в частоте покупок по крайней мере между двумя потребительскими сегментами (с учетом поправки на множественные сравнения).")
        else:
            print("\nНе удается отвергнуть нулевую гипотезу.")
            print("Статистически значимой разницы в частоте покупок между клиентскими сегментами нет (с учетом поправки на множественные сравнения).")
else:
    print("\nТест ANOVA не может быть проведен из-за ненормального распределения. Рассмотрите непараметрический тест (например, Kruskal-Wallis).")

# 5. Альтернатива: Kruskal-Wallis test (если данные не нормально распределены)
if not normal_clusters:
    # Проверка на достаточное количество кластеров
    if len(customer_data['cluster'].unique()) < 2:
        print("Warning: Less than two clusters available. Kruskal-Wallis cannot be performed.")
    else:
        # Выполняем тест Kruskal-Wallis
        h_statistic, p_value = stats.kruskal(*cluster_frequencies)
        print("\nРезультаты теста Kruskal-Wallis:")
        print(f"Kruskal-Wallis statistic: {h_statistic:.3f}")
        print(f"P-value: {p_value:.3f}")

        # Поправка Бонферрони
        alpha_corrected = alpha / (len(customer_data['cluster'].unique()) - 1)  # Bonferroni
        print(f"\nСкорректированный уровень значимости (Bonferroni): {alpha_corrected:.3f}")

        if p_value < alpha_corrected:
            print("\nОтвергаем нулевую гипотезу.")
            print("Существуют статистически значимые различия в частоте покупок между клиентскими сегментами (с учетом поправки на множественные сравнения).")
        else:
            print("\nНе удается отвергнуть нулевую гипотезу.")
            print("Статистически значимых различий в частоте покупок между клиентскими сегментами нет (с учетом поправки на множественные сравнения).")
```
